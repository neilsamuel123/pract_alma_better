```python
# To get requesite columns
data=airbnb_data[['id', 'name', 'host_id', 'host_name', 'neighbourhood_group','neighbourhood', 'room_type', 'price','minimum_nights', 'number_of_reviews',
             'last_review','reviews_per_month', 'calculated_host_listings_count','availability_365']]
```

```python
# To check the first five rows of the dataset
data['total_price']=data['minimum_nights']*data['price']
data.head()
```

```python
data['neighbourhood_group'].value_counts()
```

```python
# Price of neighbourhood group
plt.figure(figsize=(15,5))
sns.barplot(x='neighbourhood_group',y='price',hue='room_type',data=data)
```

```python
brooklyn_data=data[data['neighbourhood_group']=='Brooklyn']
brooklyn_data.head()
```

```python
# Total hosts in brooklyn and their properties
brooklyn_hosts=brooklyn_data.groupby(['host_name','room_type'])['calculated_host_listings_count'].max().astype(int).reset_index()
brooklyn_hosts.sort_values(by ='calculated_host_listings_count',ascending=False,inplace=True)
top10_hosts_brooklyn=brooklyn_hosts.head(10)
top10_hosts_brooklyn
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='host_name',y='calculated_host_listings_count',hue='room_type',data=top10_hosts_brooklyn)

```

```python
# Total hosts in brooklyn and their properties
brooklyn_likeable_hosts=brooklyn_data.groupby(['host_name','room_type'])['number_of_reviews'].max().astype(int).reset_index()
brooklyn_likeable_hosts.sort_values(by ='number_of_reviews',ascending=False,inplace=True)
top10_hosts_likeable_brooklyn=brooklyn_likeable_hosts.head(10)
top10_hosts_likeable_brooklyn
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='host_name',y='number_of_reviews',hue='room_type',data=top10_hosts_likeable_brooklyn)
```
